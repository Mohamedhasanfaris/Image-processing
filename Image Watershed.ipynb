```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading Image
img = cv2.imread("ani.jpg")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img2)
```

```python
#Grayscale
img_g = cv2.imread("ani.jpg",0)
img3 = cv2.cvtColor(img_g, cv2.COLOR_BGR2RGB)

plt.imshow(img3)
```

### image Thresholding

```python
img_1 = cv2.medianBlur(img_g,3)
# ret,th1 = cv2.threshold(img_1,100,255,cv2.THRESH_BINARY)
adpt = cv2.adaptiveThreshold(img_g,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,101,5) 
plt.imshow(adpt)
```
